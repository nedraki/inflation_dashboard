```python
import boto3
import os
from data_loader_s3 import upload_file
```

```python
# Connects to s3
s3 = boto3.resource("s3")
# Print out bucket names:
for bucket in s3.buckets.all():
    print(bucket.name)
bucket_name = bucket.name
```

```python
# Create client for session
client = boto3.client("s3")
# Metadata associated to bucket
response = client.list_buckets()
```

```python
# Retrieve bucket name
response["Buckets"]
```

```python
# Objects in Bucket:
objects_in_bucket = client.list_objects(
    Bucket=bucket_name)

objects_in_bucket
```

```python
# Display available folders in bucket
content_bucket = objects_in_bucket["Contents"]
# Loop over content
directories = [element["Key"] for element in content_bucket]
print(directories)
```
